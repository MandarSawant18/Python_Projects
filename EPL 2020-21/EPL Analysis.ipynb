```python
epl['MinsPerGame'] = (epl['Mins']/epl['Matches']).astype(int)
epl['GoalsPerGame']= (epl['Goals']/epl['Matches']).astype(float)
epl
```

```python
#Total Goals Scored in the league
total_goals = epl['Goals'].sum()
total_goals
```

```python
#Penalty Goals
total_pens = epl['Penalty_Goals'].sum()
total_pens
```

```python
#Penalties Attempted
total_pens_attempted = epl['Penalty_Attempted'].sum()
total_pens_attempted
```

```python
#Pie Chart to show pens scored vs missed
pens_not_scored = total_pens_attempted - total_pens
pens_not_scored
pie_data=[total_pens,pens_not_scored]
labels=['Penalties Scored', 'Penalties Missed']
colors=sns.color_palette('tab10')[2:4]

plt.figure(figsize=(12,5),dpi=90)
plt.pie(x=pie_data,labels=labels,colors= colors,startangle=60,labeldistance=1.1,autopct='%.0f%%')
plt.title("Penalties Converted VS Missed");
```

```python
#Top 5 countries with most players
epl['Nationality'].value_counts()[0:10]
```

```python
nations = epl.groupby('Nationality').size().sort_values(ascending=False)
top_ten_countries = nations.head(10)

fig = top_ten_countries.plot(kind='bar',figsize=(12,6),color =sns.color_palette('tab10'))
for p in fig.patches:
       fig.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+p.get_width()/2, p.get_height()+0.9),xytext=(0,4),ha ='center', va='center', textcoords='offset points')
plt.ylabel('Count of Players')
plt.title('Top 10 Countries with Max Players')
```

```python
#Clubs with Most Players
most_players_clubwise = epl['Club'].value_counts().nlargest(5)
fig = most_players_clubwise.plot(kind='bar',figsize=(12,6),color=sns.color_palette('Set2'))
for p in fig.patches:
       fig.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+p.get_width()/2, p.get_height()+0.2),xytext=(0,4),ha ='center', va='center', textcoords='offset points')
plt.ylabel('Count of Players')
plt.title('Top 5 Clubs with Max Players')
plt.xticks(rotation=50,fontsize=12);

```

```python
#Clubs with Least Players
least_players_clubwise = epl['Club'].value_counts().nsmallest(5)
fig = least_players_clubwise.plot(kind='bar',figsize=(12,6),color=sns.color_palette('Set2'))
for p in fig.patches:
       fig.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+p.get_width()/2, p.get_height()+0.2),xytext=(0,4),ha ='center', va='center', textcoords='offset points')
plt.ylabel('Count of Players')
plt.title('Top 5 Clubs with Least Players')
plt.xticks(rotation=50,fontsize=12);
```

```python
#Age wise Player Distribution

Under20 = epl[epl['Age']<=20]
Between20_25 =epl[(epl['Age']>20) & (epl['Age']<=25)]
Between25_30 =epl[(epl['Age']>25) & (epl['Age']<=30)]
Above30 = epl[epl['Age']>30]
```

```python
plt.figure(figsize=(5,5),dpi=100)
pie_data2 = np.array([Under20['Name'].count(),Between20_25['Name'].count(),Between25_30['Name'].count(),Above30['Name'].count()])
labels2 = ['20 or below','21-25','26-30','Above 30']
colors = ['#56a0d3', '#d4af37', '#006b3c', '#7c0a02']
plt.pie(x=pie_data2,labels=labels2,startangle=90,shadow= True,autopct='%.0f%%',colors = colors,radius=0.8)
plt.title("Age Wise Distribution of Players in EPL", fontsize=8);
```

```python
#Total Under 20 Players in Each Club
players_under_20 = epl[epl['Age']<20]
plt.figure(figsize=(10,4),dpi=100)
fig = players_under_20['Club'].value_counts().plot(kind='bar',color = sns.color_palette('tab10'))
plt.title('Clubwise U-20 Player Count', fontsize=12)
plt.xlabel('Teams')
plt.ylabel('No of Players')
plt.xticks(rotation =90)
for p in fig.patches:
       fig.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+p.get_width()/2, p.get_height()),xytext=(0,4),ha ='center', va='center', textcoords='offset points')
```

```python
#Total Assists From Each Club
plt.figure(figsize=(15,7),dpi=120)
fig = epl.groupby('Club').sum()['Assists'].sort_values(ascending=True).plot(kind='bar', color= sns.color_palette('tab10'))
plt.ylabel("Assists", fontsize=12)
plt.xlabel("Clubs", fontsize=12)
plt.title("Clubwise Assists", fontsize = 15)
for p in fig.patches:
       fig.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+p.get_width()/2, p.get_height()+0.2),xytext=(0,4),ha ='center', va='center', textcoords='offset points')
```

```python
#Goals with and without assists
total_assists = epl['Assists'].sum()
goals_without_assists = total_goals-total_assists

plt.figure(figsize=(10,4),dpi=100)
plt.title('Distribution of Goals')
pie_data3 = [total_assists, goals_without_assists]
labels3 = ['Goals with Assists', 'Goals without Assists']
plt.pie(x=pie_data3, labels = labels3, startangle =90, autopct='%.0f%%');
```
