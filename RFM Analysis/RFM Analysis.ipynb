```python
#Convert Order Date to datetime

df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
#Converting Customer ID into string format to be used for group by later

df["_CustomerID"] = df["_CustomerID"].apply(str)
```

```python
#The datatypes have been successfully changed

df.dtypes
```

```python
#Calculating the revenue 


df['Revenue'] = ((1-df["Discount Applied"])*df['Unit Price']) * df['Order Quantity']
```

```python
df.head()
```

```python
# Seperating the dataframe to only include columns that we want

columns = ["OrderNumber","OrderDate","_CustomerID","Revenue"]
sales_comp = df[columns]
sales_comp.head()
```

```python
# Checking the size of the dataframe
sales_comp.shape
```

```python
# min(sales_comp['OrderDate']) = Timestamp('2018-05-31 00:00:00')

# max (sales_comp['OrderDate']) = Timestamp('2020-12-30 00:00:00')


# Using the latest date as 1st Jan 2021 to avoid big numbers in calculations of recency value


today_date = pd.to_datetime('2021-01-01')
```

```python
# Calculating the recency, frequency and monetary values


rfm_dataset = sales_comp.groupby('_CustomerID').agg({
            "OrderDate": lambda v: (today_date - v.max()).days,
            "OrderNumber": 'count',
            "Revenue": 'sum'})


rfm_dataset
```

```python
#Checking the size of rfm dataset

rfm_dataset.shape
```

```python
#Checking the count of unique customers and hereby verifying the shape of rfm dataset

len(df['_CustomerID'].unique())
```

```python
#Renaming the columns of rfm_dataset

rfm_dataset = rfm_dataset.rename(
            columns={
                "OrderDate":'recency',
                "OrderNumber":'frequency',
                "Revenue":'monetary'})

rfm_dataset
```

```python
r = pd.qcut(rfm_dataset['recency'],q=5,labels=range(5,0,-1))
f = pd.qcut(rfm_dataset['frequency'],q=5,labels=range(1,6))
m = pd.qcut(rfm_dataset['monetary'],q=5,labels=range(1,6))


#For r we use reverse list because lower the number of days in recency, better for the business because it shows the customers latest purchase
```

```python
rfm = rfm_dataset.assign(R=r.values,F=f.values,M=m.values)
rfm
```

```python
rfm['rfm_group'] = rfm[['R','F','M']].apply(lambda v: '-'.join(v.astype(str)),axis=1)
rfm['rfm_score'] =  rfm[['R','F','M']].sum(axis=1)

rfm
```

```python
#Finding out core customers

rfm[rfm['rfm_score'] == 15]
```

```python
#Loyal Customers


rfm[rfm['F'] >= 3]
#len(rfm[rfm['F'] >= 3])
```

```python
#Whales


rfm[rfm['M'] >= 3]
#len(rfm[rfm['M'] >= 3])
```

```python
#Customers at risk


rfm[(rfm['R']==2) & (rfm['F'] >= 3) & (rfm['M'] >= 3)]
#len(rfm[(rfm['R']==2) & (rfm['F'] >= 3) & (rfm['M'] >= 3)])
```
